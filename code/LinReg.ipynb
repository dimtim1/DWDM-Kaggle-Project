```python
# conventional way to import pandas
import pandas as pd
```

```python
# read the training data & display first five records
data = pd.read_csv('train.csv', index_col=0)
# data.head()

# read the test data
test_data = pd.read_csv('test.csv', index_col=0)
```

```python
# check for the no of records & features
data.shape
```

```python
# select the 37 numeric and ordinal features in both training & testing dataset
num_data = data.select_dtypes(['int64', 'float64'])
# print num_data

test_num_data = test_data.select_dtypes(['int64', 'float64'])
```

```python
# fill the missing values in the data with respective column's mean
num_data.fillna(num_data.mean(), inplace=True)
# print num_data

# same thing in test dataset
test_num_data.fillna(test_num_data.mean(), inplace=True)
```

```python
# preparing feature matrix (X) and response vector (y)
feature_cols = set(num_data.columns)
feature_cols.remove('SalePrice')
feature_cols = list(feature_cols)
X =  num_data[feature_cols]
# X = num_data.ix[:, num_data.columns != 'SalePrice']
print X.shape
y = num_data.SalePrice
print y.shape
```

```python
# from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import cross_val_score
```

```python
# import Linear Regression model
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# build model, learn coefficients and evaluate it
scores = cross_val_score(linreg, X, y, cv=10, scoring='mean_squared_error')
print(scores)
```

```python
# fix the sign of MSE scores
mse_scores = -scores
print(mse_scores)
```

```python
import numpy as np
```

```python
# convert from MSE to RMSE
rmse_scores = np.sqrt(mse_scores)
print(rmse_scores)

# calculate the average RMSE
print(rmse_scores.mean())
```

```python
# create an another instantiation
linear = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X, y)

# make predictions on the testing set 
y_pred = linreg.predict(test_num_data)

# fixme : the indexing mismatch
# pd.concat([y, pd.DataFrame(y_pred)], axis=1, ignore_index=True)

# convert numpy array into panda series
print pd.DataFrame(y_pred)
```
