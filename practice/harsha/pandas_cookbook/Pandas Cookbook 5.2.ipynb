#Seaborn

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn settings
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8,4)})
```

```python
titanic = pd.read_csv('titanic.csv')
server = pd.read_csv('serverdata.csv')
mlb = pd.read_csv('mlbsalaries.csv')
```

**Histograms**

```python
plt.hist(titanic.Age.dropna(),bins=20)
```

```python
plt.hist(server.Time,25,color=sns.desaturate("indianred",1))

```

```python
plt.hist(mlb.Salary.dropna(),25)
```

**Box Plot **

```python
sns.boxplot(titanic.Age.dropna())
```

```python
sns.boxplot(titanic.Age.dropna(),titanic.Sex.dropna())
```

```python
sns.boxplot(server.Time)
```

```python
sns.boxplot(mlb.Salary)
```

** Kernel Density Estimate Plot **

```python
sns.kdeplot(titanic.Age.dropna(),shade=True)
```

```python
sns.kdeplot(server.Time.dropna(),shade=True)
```

```python
sns.kdeplot(mlb.Salary.dropna(),shade=True)
```

```python
sns.distplot(titanic.Age.dropna())
```

```python
sns.distplot(server.Time.dropna())
```

```python
sns.distplot(mlb.Salary.dropna())
```

```python
sns.violinplot(titanic.Age.dropna())
```

```python
sns.violinplot(titanic.Age.dropna(),titanic.Sex)
```

```python
sns.violinplot(mlb.Salary.dropna())
```

** Cumulative Distribution Plot **

```python
sns.kdeplot(titanic.Age.dropna(),shade=True,cumulative=True)
```

```python
sns.kdeplot(server.Time.dropna(),shade=True,cumulative=True)
```

```python
sns.kdeplot(mlb.Salary.dropna(),shade=True,cumulative=True)
```
