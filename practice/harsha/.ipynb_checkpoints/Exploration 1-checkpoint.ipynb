```python
%config InlineBackend.figure_format = 'svg' #set 'png' here when working on notebook
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import skew
from sklearn.cross_validation import cross_val_score



def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y, scoring="mean_squared_error", cv = 5))
    return(rmse)
```

```python
train = pd.read_csv("../../data/train.csv")
test = pd.read_csv("../../data/test.csv")
```

```python
train.head()
```

```python
all_data = pd.concat((train.loc[:,'MSSubClass':'SaleCondition'],
                      test.loc[:,'MSSubClass':'SaleCondition']))


numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index
categoric_feats = all_data.dtypes[all_data.dtypes == "object"].index

print 
```

```python
list(categoric_feats.values)
```

```python
all_data = pd.get_dummies(all_data)
all_data = all_data.fillna(all_data.mean())
X_train = all_data[:train.shape[0]]
X_test = all_data[train.shape[0]:]
y = train.SalePrice
```

```python
feats = list(X_train.columns.values)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()


rmse_cv(lr).mean()
```

```python
lr.fit(X_train,y)
zip(X_train.columns.values,lr.coef_)
```

```python
from sklearn.ensemble import RandomForestRegressor


rf = RandomForestRegressor()
rf.fit(X_train, y)
print "Features sorted by their score:"
tup_list = sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), feats), 
             reverse=True)

for v,feat in tup_list:
    print feat,v
```

# Recursive Elimination

```python
from sklearn.feature_selection import RFE

lr = LinearRegression()
#rank all features, i.e continue the elimination until the last one
rfe = RFE(lr, n_features_to_select=1)
rfe.fit(X_train,y)
 
print "Features sorted by their rank:"
tup_list =  sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), feats))

for v,feat in tup_list:
    print v,feat
```

```python
top_feat = [feat for rank,feat in tup_list][200:]
```

```python
X_new = X_train[top_feat]
```

```python
X_new.head()
```

```python
lr = LinearRegression()
lr.fit(X_new,y)
```

```python
print np.sqrt(-cross_val_score(lr, X_new, y, scoring="mean_squared_error", cv = 5)).mean()
```

```python
rf = RandomForestRegressor(n_estimators=100,max_depth=10)
```

```python
rf.fit(X_train,y)
 
print "Features sorted by their rank:"
tup_list =  sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), feats),reverse=True)

for v,feat in tup_list:
    print v,feat
```

```python
top_feats = [feat for score,feat in tup_list if score > 0.002]
```

```python
top_feats
```

```python
X_new = X_train[top_feats]
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_new,y)
print np.sqrt(-cross_val_score(rf, X_train, y, scoring="mean_squared_error", cv = 5)).mean()
```
